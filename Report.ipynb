# 3) description of the algorithm

The code is structured around a State-value function, which is in face a big dictionnary : 
- Each key is a description of a met state : there are 37 parameters.
- To each key is linked an array wich contains the expected reward for each action possible. Here there are 4 actions, so the array is 4 in lenght. 

The environement generate 300 steps for one episode. 

At the first episode the actions are selected randomly, and the reward (at the end of the episod) for each step (an entry in the dictionnary) is saved in the associated cell of the array.  
Then at the second episod, at each state we generate slight advantage for the action who have been giving the highest reward, but still choose randomly, but not with equiprobable randomness... Then updating the action-value with the consequent reward.  
Repeating this process 3000 episodes, over 300 step for each episode, gives 900 000 possibility to improve the State-Value fonction


# 4) Solving the problem

Our implementation solve the problem as in 71 steps the agent get 13 points.
